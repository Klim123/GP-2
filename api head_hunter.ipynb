```python
import requests
import pandas as pd

# URL API HeadHunter для поиска вакансий
url = "https://api.hh.ru/vacancies"

vacancies_list = []

count = 0

for i in range(100):
    print(i)
    # Параметры запроса
    params = {
        "per_page": 50,  # Количество вакансий на страницу
        "page": i
        }
    
    # Отправка GET-запроса
    response = requests.get(url, params=params)
    
    # Проверяем статус-код
    if response.status_code == 200:
        count+=1
        vacancies = response.json()
        for vacancy in vacancies["items"]:
            vacancy_url = vacancy['url']
            try:
                vacancy_response = requests.get(vacancy_url)
                if vacancy_response.status_code == 200:
                    count+=1
                    vacancy["info from url"] = vacancy_response.json()
                else:
                    vacancy["info from url"] = {}
            except Exception as e:
                vacancy["info from url"] = {}
            vacancies_list.append(vacancy)
    else:
        print(f"Окончание расчётов")
        print(f"Обработано {i} страниц")
        print(f"Сделано {count} api запросов")
        break

print(f"Количество вакансий {len(vacancies_list)}")
df = pd.DataFrame(vacancies_list)
df.to_csv("vacancies_hh.csv")

import json
with open("vacancies_hh.json", "w", encoding="utf-8") as file:
    json.dump(vacancies_list, file, indent=4, ensure_ascii=False)
```
