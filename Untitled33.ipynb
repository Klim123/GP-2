```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
avito = pd.read_csv('all_vacancies_avito.csv')
```

```python
avito
```

```python
rating=avito['Местоположение'].value_counts()
colors = sns.color_palette('pastel')[0:5]
small = rating[rating / rating.sum() < 0.05]
rating2 = rating[rating / rating.sum() >= 0.05]
rating2['Other'] = small.sum()
sizes=rating2.values/rating2.values.sum()
plt.pie(sizes,labels=rating2.index,autopct='%.1f%%',textprops={'fontfamily':'monospace'}, colors=colors)
plt.title('В каких городах находится на avito',fontfamily='monospace',fontweight='bold')
plt.show()
```

```python
rating=avito['Зарплата'].value_counts()
colors = sns.color_palette('pastel')[0:5]
small = rating[rating / rating.sum() < 0.05]
rating2 = rating[rating / rating.sum() >= 0.05]
rating2['Other'] = small.sum()
sizes=rating2.values/rating2.values.sum()
plt.pie(sizes,labels=rating2.index,autopct='%.1f%%',textprops={'fontfamily':'monospace'}, colors=colors)
plt.title('В какой валюте предлагают зарплату на avito',fontfamily='monospace',fontweight='bold')
plt.show()
```

```python
top_cities = avito["Местоположение"].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette="viridis")
plt.xlabel("Количество вакансий")
plt.ylabel("Город")
plt.title("Топ-10 городов по количеству вакансий")
plt.show()
```

Вывод:

Наибольшее количество вакансий сосредоточено в крупных мегаполисах (Москва, Санкт-Петербург, Казань и другие).
Это логично, так как в больших городах выше спрос на работников и развита экономика.
В регионах количество вакансий значительно ниже, что говорит о меньшем рынке труда.
Что это значит для бизнеса?

В крупных городах конкуренция за сотрудников выше, поэтому компании могут предлагать лучшие условия.
В небольших городах можно расширять присутствие, если там мало конкурентов.

Распределение условий работы

```python
top_conditions = avito["Условия"].value_counts().head(5)
plt.figure(figsize=(8, 8))
plt.pie(top_conditions, labels=top_conditions.index, autopct="%1.1f%%", colors=sns.color_palette("viridis"))
plt.title("Распределение условий работы (топ-5)")
plt.show()
```

Вывод:

Гибкий график и сменный график являются самыми популярными условиями работы.
Вахтовый метод встречается реже, но он может предлагать более высокие зарплаты.
Что это значит для бизнеса?

Если гибкий график популярен, можно сделать его стандартом в вакансиях.
Если вахта дает высокие зарплаты, можно привлекать работников из регионов.

```python
top_jobs = avito["Название вакансии"].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette="viridis")
plt.xlabel("Количество вакансий")
plt.ylabel("Название вакансии")
plt.title("Топ-10 популярных вакансий на Avito")
plt.show()
```

Это логично, так как в больших городах выше спрос на работников и развита экономика.
В регионах количество вакансий значительно ниже, что говорит о меньшем рынке труда.
Что это значит для бизнеса?

В крупных городах конкуренция за сотрудников выше, поэтому компании могут предлагать лучшие условия.
В небольших городах можно расширять присутствие, если там мало конкурентов.

```python
courier_df = avito[avito["Название вакансии"].str.contains("курьер в магазин", case=False, na=False)]
courier_salary_city = courier_df.groupby("Местоположение")["Зарплата"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=courier_salary_city.head(10).values, y=courier_salary_city.head(10).index, palette="viridis")
plt.xlabel("Средняя зарплата")
plt.ylabel("Город")
plt.title("Топ-10 городов с самой высокой зарплатой курьеров в магазин")
plt.show()
```

Вывод:

В некоторых городах зарплата курьеров выше, чем в других.
Скорее всего, это зависит от уровня жизни в городе, конкуренции на рынке труда и спроса на доставку.
Что это значит для бизнеса?

Если в городе низкая зарплата курьеров, возможно, там не хватает вакансий.
В городах с высокой зарплатой можно протестировать новые форматы доставки (например, премиальные курьеры, экспресс-доставка).

```python
city_salary = avito.groupby("Местоположение")["Зарплата"].mean().sort_values(ascending=False)
top_cities_highest = city_salary.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities_highest.values, y=top_cities_highest.index, palette="viridis")
plt.xlabel("Средняя зарплата")
plt.ylabel("Город")
plt.title("Топ-10 городов с самой высокой средней зарплатой")
plt.show()
```

```python
conditions_salary = avito.groupby("Условия")["Зарплата"].mean().sort_values(ascending=False)
top_conditions = conditions_salary.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_conditions.values, y=top_conditions.index, palette="viridis")
plt.xlabel("Средняя зарплата (RUB)")
plt.ylabel("Условия работы")
plt.title("Топ-10 условий работы с самой высокой зарплатой")
plt.show()
```

```python
file_path = "vacancies_hh.csv"
df = pd.read_csv(file_path)
with open('vacancies_hh.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
df = pd.DataFrame(data)
df['city'] = df['area'].apply(lambda x: x['name'])
df['salary_from'] = df['salary'].apply(lambda x: x.get('from', None) if x is not None else None)
df['salary_to'] = df['salary'].apply(lambda x: x.get('to', None) if x is not None else None)
df['salary_currency'] = df['salary'].apply(lambda x: x.get('currency', None) if x is not None else None)
df['type_name'] = df['type'].apply(lambda x: x['name'])
df["salary_from"] = pd.to_numeric(df["salary_from"], errors="coerce")
df["salary_to"] = pd.to_numeric(df["salary_to"], errors="coerce")

```

```python
df.columns
```

```python
df=df[['name','city','salary_from','salary_to','salary_currency','snippet','employment_form','experience']]
```

```python
df['employment_form'] = df['employment_form'].apply(lambda x: x.get('name', None) if x is not None else None)
df['Описание'] = df['snippet'].apply(lambda x: x.get('requirement', None) if x is not None else None)
df['experience'] = df['experience'].apply(lambda x: x.get('name', None) if x is not None else None)

df=df.drop('employment_form',axis=1)
df=df.drop('snippet',axis=1)
```

```python
# Преобразуем зарплаты в числовой формат
df["salary_from"] = pd.to_numeric(df["salary_from"], errors="coerce")
df["salary_to"] = pd.to_numeric(df["salary_to"], errors="coerce")
```

```python
df
```

```python
df['salary_currency'].value_counts()
```

```python
exchange_rates = {
    'RUR': 1,
    'KZT': 0.18,
    'BYR': 27.77,
    'UZS': 0.00692,
    'USD': 89.24,
    'KGS': 1.02,
    'EUR': 93.9,
    'GEL': 31.98,
    'AZN': 52.5
}
df['salary_from_rur'] = df['salary_from'] * df['salary_currency'].map(exchange_rates)
df['salary_to_rur'] = df['salary_to'] * df['salary_currency'].map(exchange_rates)
```

```python
rating=df['city'].value_counts()
colors = sns.color_palette('pastel')[0:5]
small = rating[rating / rating.sum() < 0.05]
rating2 = rating[rating / rating.sum() >= 0.05]
rating2['Other'] = small.sum()
sizes=rating2.values/rating2.values.sum()
plt.pie(sizes,labels=rating2.index,autopct='%.1f%%',textprops={'fontfamily':'monospace'}, colors=colors)
plt.title('В каких городах находится на hh.ru',fontfamily='monospace',fontweight='bold')
plt.show()
```

```python
rating=df['salary_currency'].value_counts()
colors = sns.color_palette('pastel')[0:5]
small = rating[rating / rating.sum() < 0.05]
rating2 = rating[rating / rating.sum() >= 0.05]
rating2['Other'] = small.sum()
sizes=rating2.values/rating2.values.sum()
plt.pie(sizes,labels=rating2.index,autopct='%.1f%%',textprops={'fontfamily':'monospace'}, colors=colors)
plt.title('В каких валютах предлагают з/п на hh.ru',fontfamily='monospace',fontweight='bold')
plt.show()
```

```python
top_cities = df["city"].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Город")
plt.ylabel("Количество вакансий")
plt.title("Топ-10 городов по количеству вакансий")
plt.grid(True)
plt.show()
```

```python
exp_salary = df.groupby("experience")["salary_from_rur"].median().dropna().sort_values()
plt.figure(figsize=(12, 6))
sns.barplot(x=exp_salary.index, y=exp_salary.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Опыт работы")
plt.ylabel("Медианная зарплата")
plt.title("Средняя зарплата в зависимости от опыта")
plt.grid(True)
plt.show()
```

```python
df_filtered = df[df["salary_from_rur"] < df["salary_from_rur"].quantile(0.95)]  # Убираем выбросы
plt.figure(figsize=(12, 6))
sns.histplot(df_filtered["salary_from_rur"].dropna(), bins=50, kde=True, color="blue")
plt.xlabel("Зарплата")
plt.ylabel("Количество вакансий")
plt.title("Распределение зарплат")
plt.grid(True)
plt.show()
```
